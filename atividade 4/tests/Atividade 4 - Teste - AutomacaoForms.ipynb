```python


import pyautogui
import time

def fill_out_form(name, cpf, email, password):
    pyautogui.write(name, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.write(cpf, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.write(email, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.write(password, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.3)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.press('enter')

def clear_form():
    time.sleep(1)

    button_z, button_w = pyautogui.locateCenterOnScreen(r"C:\Users\joao.aureliano\Downloads\projects-testes-main(1)\projects-testes-main\atividade 4\tests\a.png") #endereço imagem neutra
    pyautogui.click(button_z, button_w)
    time.sleep(1)
    
    button_x, button_y = pyautogui.locateCenterOnScreen(r"C:\Users\joao.aureliano\Downloads\projects-testes-main(1)\projects-testes-main\atividade 4\tests\botao de apagar.png") #endereço do botao apagar

    pyautogui.click(button_x, button_y)
    
    time.sleep(0.5)
    button_z, button_w = pyautogui.locateCenterOnScreen(r"C:\Users\joao.aureliano\Downloads\projects-testes-main(1)\projects-testes-main\atividade 4\tests\a.png") #endereço imagem neutra
    time.sleep(0.5)
    
    button_z, button_w = pyautogui.locateCenterOnScreen(r"C:\Users\joao.aureliano\Downloads\projects-testes-main(1)\projects-testes-main\atividade 4\tests\botao de nome.png") #endereço do botao nome

    pyautogui.click(button_z, button_w)

def test_valido():
    print("teste com todas as entidades validas")
    fill_out_form("Teste com informacoes corretas", "11111111111", "Meu@Email.com", "senha123")
    print("sucesso")

def test_nome():
    print("teste com nome vazio")
    fill_out_form("", "11111111111", "Meu@Email.com", "senha123")
    time.sleep(1)
    print("sucesso")

    clear_form()

    print("teste com nome valido")
    fill_out_form("Meu nome", "12345678909", "Meu@Email.com", "senha123")
    time.sleep(1)
    pyautogui.press('enter')
    time.sleep(1)
    print("sucesso")

    clear_form()

def test_cpf():
    print("teste com cpf vazio")
    fill_out_form("Meu nome", "", "MeuEmail@exemplo.com", "senha123")
    time.sleep(1)
    print("sucesso")

    clear_form()

    print("teste com cpf invalido")
    fill_out_form("Meu nome", "12345678910", "MeuEmail@exemplo.com", "senha123")
    time.sleep(0.5)
    pyautogui.press('enter')
    time.sleep(1)
    print("sucesso")

    clear_form()

    print("teste com cpf valido")
    fill_out_form("Meu nome", "12345678909", "MeuEmail@exemplo.com", "senha123")
    time.sleep(0.5)
    pyautogui.press('enter')
    time.sleep(1)
    print("sucesso")

    clear_form()

def test_email():
    print("teste com email vazio")
    fill_out_form("Meu nome", "12345678909", "", "senha123")
    time.sleep(1)
    print("sucesso")

    clear_form()

    print("teste com email invalido")
    fill_out_form("Meu nome", "12345678909", "MeuEmailInvalidoSemArroba", "senha123")
    time.sleep(1)
    print("sucesso")

    clear_form()

    print("teste com email valido")
    fill_out_form("Meu nome", "12345678909", "MeuEmail@exemplo.com", "senha123")
    time.sleep(0.5)
    pyautogui.press('enter')
    time.sleep(1)
    print("sucesso")

    clear_form()

def test_senha():
    print("teste com senha vazia")
    fill_out_form("Meu nome", "12345678909", "MeuEmail@exemplo.com", "")
    time.sleep(1)
    print("sucesso")

    clear_form()

    print("teste com senha curta")
    fill_out_form("Meu nome", "12345678909", "MeuEmail@exemplo.com", "senha")
    time.sleep(1)
    print("sucesso")
    
    clear_form()

    print("teste com senha valida")
    fill_out_form("Meu nome", "12345678909", "MeuEmail@exemplo.com", "senha123")
    time.sleep(0.5)
    pyautogui.press('enter')
    time.sleep(1)
    print("sucesso")

    clear_form()

pyautogui.hotkey('alt', 'tab')
time.sleep(0.5)
pyautogui.press('tab')
time.sleep(1)
test_nome()
test_cpf()
test_email()
test_senha()
test_valido()
```
