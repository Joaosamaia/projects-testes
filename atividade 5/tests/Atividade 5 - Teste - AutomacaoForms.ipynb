```python
import pyautogui
import time
import threading

def fill_out_form(name, cpf, email, password):
    pyautogui.typewrite(name, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(cpf, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(email, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(password, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.3)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.press('enter')

def clear_form():
    time.sleep(1)

    button_z, button_w = pyautogui.locateCenterOnScreen(r"./a.png") #endereço imagem neutra
    pyautogui.click(button_z, button_w)
    time.sleep(1)
    
    button_x, button_y = pyautogui.locateCenterOnScreen(r";/botao de apagar.png") #endereço do botao apagar

    pyautogui.click(button_x, button_y)
    
    time.sleep(0.5)
    button_z, button_w = pyautogui.locateCenterOnScreen(r"./a.png") #endereço imagem neutra
    time.sleep(0.5)
    
    button_z, button_w = pyautogui.locateCenterOnScreen(r"./botao de nome.png") #endereço do botao nome

    pyautogui.click(button_z, button_w)

def test_valido_form1():
    print("teste preenchendo o formulario 2")

    pyautogui.hotkey('alt', 'tab')
    time.sleep(0.5)
    pyautogui.press('tab')
    time.sleep(1)

    fill_out_form("Teste com informacoes corretas", "11111111111", "Meu@Email.com", "senha123")
    print("sucesso")

def test_valido_form2():
    print("teste preenchendo o formulario 1")
    pyautogui.hotkey('alt', 'tab')
    time.sleep(0.5)
    pyautogui.press('tab')
    time.sleep(1)
    
    fill_out_form("Teste com informacoes corretas", "11111111111", "Meu@Email.com", "senha123")
    print("sucesso")

thread1 = threading.Thread(target=test_valido_form1)
thread2 = threading.Thread(target=test_valido_form2)

thread1.start()
thread2.start()

thread1.join()
thread2.join()
```
