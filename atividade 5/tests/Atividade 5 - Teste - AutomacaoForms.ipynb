```python


import pyautogui
import time
import threading

def fill_out_form_1(titulo_janela, login, name, cpf, password, rua, numero, bairro, cidade, email):

    pyautogui.getWindowsWithTitle(titulo_janela)[0].activate()

    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(login, interval=0.1)
    time.sleep(1)
    pyautogui.typewrite(name, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(cpf, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(password, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(rua, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(numero, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(bairro, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(cidade, interval=0.1)
    time.sleep(1)

    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.press('down', presses=5,  interval=0.25)
    time.sleep(1)

    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.press('right')
    time.sleep(0.5)
    pyautogui.press('space')
    time.sleep(0.5)
    pyautogui.press('tab')
    time.sleep(0.5)
    
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(email, interval=0.1)
    time.sleep(1)
    
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.press('space')
    time.sleep(0.5)
    
    pyautogui.press('tab')
    time.sleep(0.3)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.press('enter')

def fill_out_form_2(titulo_janela, login, name, cpf, password, rua, numero, bairro, cidade, email):

    pyautogui.getWindowsWithTitle(titulo_janela)[0].activate()

    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(login, interval=0.1)
    time.sleep(1)
    pyautogui.typewrite(name, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(cpf, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(password, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(rua, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(numero, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(bairro, interval=0.1)
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(cidade, interval=0.1)
    time.sleep(1)

    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.press('down', presses=5,  interval=0.25)
    time.sleep(1)

    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.press('space')
    time.sleep(0.5)
    pyautogui.press('tab')
    time.sleep(0.5)
    
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.typewrite(email, interval=0.1)
    time.sleep(1)
    
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.press('space')
    time.sleep(0.5)
    
    pyautogui.press('tab')
    time.sleep(0.3)
    pyautogui.press('tab')
    time.sleep(0.5)
    pyautogui.press('enter')

def test_valido_form1():
    time.sleep(0.5)
    try:
        time.sleep(0.5)
        print("teste preenchendo o formulario 1")
        time.sleep(0.5)
    
        fill_out_form_1("Formulario 1 Teste", "Meu login", "Meu nome", "11111111111", "senha123", "Minha rua", "Meu numero", "Meu bairro", "Minha cidade", "Meu@Email.com")
        print("Sucesso ao preencher o formulario 1")
    except TypeError:
        print("Form 1 não preenchido.")

def test_valido_form2():
    time.sleep(1)
    try:
        time.sleep(0.5)
        print("teste preenchendo o formulario 2")
        time.sleep(0.5)
    
        fill_out_form_2("Formulario 2 Teste", "Meu outro login", "Meu outro nome", "11111111111", "senha123", "Minha outra rua", "Meu outro numero", "Meu outro bairro", "Minha outra cidade", "MeuOutro@Email.com")
        print("sucesso ao preencher o formulario 2")
        
    except TypeError:
        print("Form 2 não preenchido.")

time.sleep(1.5)
pyautogui.hotkey('alt', 'tab')
time.sleep(1.5)

thread1 = threading.Thread(target=test_valido_form1)
thread2 = threading.Thread(target=test_valido_form2)

thread1.start()

thread1.join()

thread2.start()

thread2.join()
```
