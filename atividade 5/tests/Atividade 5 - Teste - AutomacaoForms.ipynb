```python
import pyautogui
import time

def switch(titulo_janela):
    pyautogui.getWindowsWithTitle(titulo_janela)[0].activate()
    pyautogui.press('tab')

def preenche (info):
    time.sleep(0.05)
    pyautogui.typewrite(info, interval=0.005)

def preencher_forms(titulo_janela, login, name, cpf, password, rua, numero, bairro, cidade, email, titulo_janela2, login2, name2, cpf2, password2, rua2, numero2, bairro2, cidade2, email2):

    switch(titulo_janela)
    preenche(login)
    switch(titulo_janela2)
    preenche(login2)

    switch(titulo_janela)
    preenche(name)
    switch(titulo_janela2)
    preenche(name2)

    switch(titulo_janela)
    preenche(cpf)
    switch(titulo_janela2)
    preenche(cpf2)

    switch(titulo_janela)
    preenche(password)
    switch(titulo_janela2)
    preenche(password2)

    switch(titulo_janela)
    preenche(rua)
    switch(titulo_janela2)
    preenche(rua2)
    
    switch(titulo_janela)
    preenche(numero)
    switch(titulo_janela2)
    preenche(numero2)

    switch(titulo_janela)
    preenche(bairro)
    switch(titulo_janela2)
    preenche(bairro2)

    switch(titulo_janela)
    preenche(cidade)
    switch(titulo_janela2)
    preenche(cidade2)

    switch(titulo_janela)
    pyautogui.press('down', presses=5,  interval=0.15)
    switch(titulo_janela2)
    pyautogui.press('up', presses=3,  interval=0.15)
    

    switch(titulo_janela)
    pyautogui.press('right')
    time.sleep(0.1)
    pyautogui.press('space')
    time.sleep(0.1)
    switch(titulo_janela2)
    pyautogui.press('right', presses=2)
    time.sleep(0.1)
    pyautogui.press('space')

    switch(titulo_janela)
    preenche(email)
    switch(titulo_janela2)
    preenche(email2)
    
    switch(titulo_janela)
    pyautogui.press('space')
    time.sleep(0.1)
    switch(titulo_janela2)
    pyautogui.press('space')
    
    switch(titulo_janela)
    pyautogui.press('tab')
    pyautogui.press('enter')
    switch(titulo_janela2)
    pyautogui.press('tab')
    pyautogui.press('enter')

try:
    print("teste preenchendo os formularios simuntaneamente")
    time.sleep(0.15)

    
    preencher_forms("Formulario 1 Teste", "Meu login", "Meu nome", "11111111111", "senha123", "Minha rua", "101", "Meu bairro", "Minha cidade", "Meu@Email.com", "Formulario 2 Teste", "Meu outro login", "Meu outro nome", "12345678909", "123senha", "Minha outra rua", "202", "Meu outro bairro", "Minha outra cidade", "MeuOutro@Email.com")
    print("Sucesso ao preencher os formularios")
    
except TypeError:
    print("Forms n√£o preenchidos.")
```
